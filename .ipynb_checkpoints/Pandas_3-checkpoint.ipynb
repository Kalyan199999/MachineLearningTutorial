# Missing Values Handling

```python
num = pd.read_csv("C:\\Users\\pavan Kalyan\\Desktop\\ML\\homeprices.csv" )
num
```

```python
num1= pd.read_csv( "C:\\Users\\pavan Kalyan\\Desktop\\ML\\homeprices.csv" , na_values = ['not available' , 'no values'] )
num1
```

```python
num2= pd.read_csv( "C:\\Users\\pavan Kalyan\\Desktop\\ML\\homeprices.csv" , na_values = {'price' : 'not available'} )
num2
```

```python
# keep_default_na = False -->shows the dataset as it is and if it is true the NA is placed for missing values
num3= pd.read_csv( "C:\\Users\\pavan Kalyan\\Desktop\\ML\\homeprices.csv" , keep_default_na = False)
num3
```

```python
# isnull --> return true if null values are present
num2.isnull()
```

```python
num2.isnull().sum()
# no.of null values in each column
```

```python
# total null values
num2.isnull().sum().sum()
```

```python
# notnull() -->opposite to isnull()
# true --> not null values
num2.notnull()
```

```python
num2.notnull().sum()
```

```python
num2.notnull().sum().sum()
```

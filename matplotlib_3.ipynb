# Bar Chart

```python
from matplotlib import pyplot as plt

import numpy as np

from matplotlib import style
```

```python
plt
```

```python
classes = ["python","c++","Java","Java Script","DS"]
class1 = [30,10,20,25,10]
class2 = [40,5,20,20,10]
class3 = [35,5,30,15,15]
```

```python
plt.bar(classes,class1,color = 'r')
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
plt.barh(classes,class1,color = 'r')
```

```python
plt.bar(classes,class1,color = 'm',width = 0.1)
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
plt.bar(classes,class1,color = 'm',align ="edge" ,width =0.3)
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
plt.bar(classes,class1,color = 'b' ,width =0.3 , edgecolor = "r")
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
plt.bar(classes,class1,color = 'b' ,width =0.3 , edgecolor = "r" ,linewidth = 4)
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
plt.bar(classes,class1,color = 'm', alpha = 0.5)
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
plt.bar(classes,class1,color = 'y', linewidth =5 ,linestyle = "-.",edgecolor = "r")
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
style.use("ggplot")
plt.bar(classes,class1, color = 'y', width =0.5 ,linewidth = 2,edgecolor = 'm')
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (15,8))

plt.bar(classes,class1, color = 'y', width =0.5 ,linewidth = 2,edgecolor = 'm')
plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (15,8))

plt.bar(classes,class1, color = 'y', width =0.2 ,linewidth = 2,edgecolor = 'm' ,label = "class1")
plt.bar(classes,class2, color = 'r', width =0.2 ,linewidth = 2,edgecolor = 'b',label = "class2")

plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (15,8))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index  ,class1, width, color = 'y'  ,linewidth = 2,edgecolor = 'm' ,label = "class1")
plt.bar(class_index + width ,class2, width,color = 'r'  ,linewidth = 2,edgecolor = 'b', label = "class2")

plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (15,8))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index  ,class1, width, color = 'y'  ,linewidth = 2,edgecolor = 'm' ,label = "class1")
plt.bar(class_index + width ,class2, width,color = 'r'  ,linewidth = 2,edgecolor = 'b', label = "class2")

plt.xticks(class_index + width ,classes)

plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (15,8))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index  ,class1, width, color = 'y'  ,linewidth = 2,edgecolor = 'm' ,label = "class1")
plt.bar(class_index + width ,class2, width,color = 'r'  ,linewidth = 2,edgecolor = 'b', label = "class2")

plt.xticks(class_index + width ,classes ,rotation = 45)

plt.title("Languages")
plt.xlabel("Programming Lang")
plt.ylabel("No.of Students")
plt.legend()
plt.show()
```

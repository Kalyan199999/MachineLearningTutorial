# Histogram

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,440,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age )
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")

plt.show()
```

```python
plt.hist(ml_student_age , rwidth = 0.6 )
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")

plt.show()
```

```python
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth = 0.8)
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")

plt.show()
```

```python
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8 ,histtype = 'bar')
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")

plt.show()
```

```python
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8 ,orientation = "horizontal" , color = "m")
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")

plt.show()
```

```python
bins = [10,15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8 ,color = "y")
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")
plt.legend("ML Students")
plt.show()
```

```python
bins = [10,15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth = 0.8 ,color = "r" , label ="ML Students")

plt.hist(py_student_age, bins, rwidth = 0.8 ,color = "m" , label ="Python Students")

plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist([ml_student_age,py_student_age], bins, rwidth = 0.8 ,color = ["r","y"] , label =["ML Students","Python Students"])


plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No.of Students")
plt.legend()
plt.show()
```

# Handling missing Values

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\pavan Kalyan\\Desktop\\ML\\homeprices.csv",na_values = ["not available" , "no values"])

df
```

```python
# drops the with NULL values
df.dropna()
```

```python
# 0-->row
# 1-->col
# if axis == 0 then return all the not null row or drops the rows if they contains at least one null value
# if axis == 1 the return all the not null col or drops the cols if they contains at least one null value
df.dropna(axis = 0)
```

```python
df.dropna(axis=1 ,how = "any")
```

```python
# if all the row values is null then it drops
df.dropna(axis=1 ,how = "all")
```

```python
df.dropna(thresh = 2)
```

```python
df.dropna(thresh =3)
```

# fillna --> fils the NaN values with given values(input)

```python
df.fillna(19)
```

```python
df.fillna({"area" : 1719 , "names":"bigil" , "price" : 9999})
```
